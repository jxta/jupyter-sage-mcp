````python
# ã‚»ãƒ«1: MCPã‚»ãƒƒãƒˆã‚¢ãƒƒãƒ—
import sys
import os

# srcãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªã‚’ãƒ‘ã‚¹ã«è¿½åŠ 
src_path = os.path.join(os.path.dirname(os.getcwd()), 'src')
if src_path not in sys.path:
    sys.path.append(src_path)

# ãƒ­ãƒ¼ã‚«ãƒ«ç’°å¢ƒã®å ´åˆ
if os.path.exists('../src'):
    sys.path.append('../src')

from mcp_sage_helper import setup_sage_mcp

# MCPã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆã®åˆæœŸåŒ–
mcp = setup_sage_mcp()

print("ğŸš€ MCP Setup Complete!")
print("ğŸ“Š System Information:")
for key, value in mcp.get_system_info().items():
    print(f"  {key}: {value}")
````

````python
# ã‚»ãƒ«2: SageMath åŸºæœ¬æ©Ÿèƒ½ãƒ†ã‚¹ãƒˆ
print("ğŸ”¬ Testing SageMath functionality...")

sage_code = """
# åŸºæœ¬çš„ãªæ•°å¼å‡¦ç†
var('x y')
f = x^2 + y^2
print("Function f(x,y) =", f)
print("âˆ‚f/âˆ‚x =", diff(f, x))
print("âˆ‚f/âˆ‚y =", diff(f, y))

# æ–¹ç¨‹å¼ã®è§£
equation = x^2 - 5*x + 6 == 0
solutions = solve(equation, x)
print("Solutions to xÂ² - 5x + 6 = 0:", solutions)

# æ•°è«–
print("Prime factorization of 1001:", factor(1001))
print("First 10 primes:", [nth_prime(i) for i in range(1, 11)])
"""

result = mcp.execute_code(sage_code, kernel_type="sagemath")
print("Result status:", result["status"])
if result["status"] == "success":
    if result.get("output"):
        print("Output:", result["output"])
    if result.get("result"):
        print("Result:", result["result"])
else:
    print("Error:", result.get("error"))
````

````python
# ã‚»ãƒ«3: Python ãƒ‡ãƒ¼ã‚¿åˆ†æã¨ãƒ“ã‚¸ãƒ¥ã‚¢ãƒ©ã‚¤ã‚¼ãƒ¼ã‚·ãƒ§ãƒ³
print("ğŸ Testing Python data analysis...")

python_code = """
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ãƒ‡ãƒ¼ã‚¿ç”Ÿæˆ
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + 0.1 * np.random.randn(100)
y2 = np.cos(x) + 0.1 * np.random.randn(100)

# DataFrameã®ä½œæˆ
df = pd.DataFrame({
    'x': x,
    'sin_noisy': y1,
    'cos_noisy': y2
})

print("DataFrame shape:", df.shape)
print("Statistics:")
print(df.describe())

# ãƒ—ãƒ­ãƒƒãƒˆ
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'b.', alpha=0.6, label='sin(x) + noise')
plt.plot(x, np.sin(x), 'r-', label='sin(x)')
plt.title('Noisy Sine Wave')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'g.', alpha=0.6, label='cos(x) + noise')
plt.plot(x, np.cos(x), 'r-', label='cos(x)')
plt.title('Noisy Cosine Wave')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print("Data analysis complete!")
"""

result = mcp.execute_code(python_code, kernel_type="python")
print("Result status:", result["status"])
if result["status"] == "success":
    if result.get("output"):
        print("Output:")
        print(result["output"])
else:
    print("Error:", result.get("error"))
````

````python
# ã‚»ãƒ«4: ã‚«ãƒ¼ãƒãƒ«åˆ‡ã‚Šæ›¿ãˆãƒ‡ãƒ¢
print("ğŸ”„ Testing kernel switching...")

# ç¾åœ¨ã®ã‚«ãƒ¼ãƒãƒ«ã‚’ç¢ºèª
print(f"Current kernel: {mcp.current_kernel}")

# SageMathã«åˆ‡ã‚Šæ›¿ãˆ
switch_result = mcp.switch_kernel_context('sagemath')
print(f"Switch to SageMath: {switch_result}")

# SageMathã§è¤‡ç´ æ•°è¨ˆç®—
complex_calc = """
# è¤‡ç´ æ•°ã¨ä»£æ•°
I = sqrt(-1)
z = 3 + 4*I
print("Complex number z =", z)
print("Magnitude |z| =", abs(z))
print("z * conjugate(z) =", z * conjugate(z))

# å¤šé …å¼
R.<x> = PolynomialRing(QQ)
p = x^3 - 2*x^2 + x - 1
print("Polynomial p(x) =", p)
print("p(2) =", p(2))
"""

result = mcp.execute_code(complex_calc, kernel_type="sagemath")
print("\nSageMath complex calculation:")
if result["status"] == "success":
    print(result.get("output", "No output"))
else:
    print("Error:", result.get("error"))

# Pythonã«åˆ‡ã‚Šæ›¿ãˆ
switch_result = mcp.switch_kernel_context('python')
print(f"\nSwitch to Python: {switch_result}")

# Pythonã§çµ±è¨ˆè¨ˆç®—
stats_calc = """
import scipy.stats as stats
import numpy as np

# æ­£è¦åˆ†å¸ƒã‹ã‚‰ã‚µãƒ³ãƒ—ãƒªãƒ³ã‚°
np.random.seed(123)
data = np.random.normal(100, 15, 1000)

print(f"Sample size: {len(data)}")
print(f"Mean: {np.mean(data):.2f}")
print(f"Std: {np.std(data):.2f}")
print(f"95% CI for mean: {stats.t.interval(0.95, len(data)-1, np.mean(data), stats.sem(data))}")
"""

result = mcp.execute_code(stats_calc, kernel_type="python")
print("\nPython statistical calculation:")
if result["status"] == "success":
    print(result.get("output", "No output"))
else:
    print("Error:", result.get("error"))
````
